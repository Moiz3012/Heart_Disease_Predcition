# üíì Heart Disease Prediction Project

**Description:**  
This notebook demonstrates a Logistic Regression model to predict the probability of heart disease based on patient features. We normalize data, train the model, and visualize predictions.

---

## 1Ô∏è‚É£ Data Loading & Exploration

- Load dataset from Excel.  
- Explore features like `age`, `sex`, `cholesterol`, `trestbps`, etc.  
- Check for missing values and basic statistics.

```python
import numpy as np
import math
import copy
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
import seaborn as sns
```

```python
df = pd.read_excel('heart_disease_risk.xlsx')
X_trains = df[['age', 'sex', 'cp', 'trestbps', 'chol', 'fbs', 'restecg', 'thalach', 'exang', 'oldpeak', 'slope', 'ca', 'thal']].values
y_trains = df['target'].values

```


## 2Ô∏è‚É£ Data Preprocessing

- Split dataset into training and test sets.  
- Normalize features using:  

```python
# split first
X_train, X_test, y_train, y_test = train_test_split(X_trains, y_trains, test_size=0.2, random_state=42)

# normalize *only using train data*
X_mean = np.mean(X_train, axis=0)
X_std = np.std(X_train, axis=0)

X_train = (X_train - X_mean) / X_std
X_test = (X_test - X_mean) / X_std
```

```python
print(X_train.shape)
print(X_test.shape)
```

```python
w = [-0.61625282, 0.04969199, -0.22803034, 0.15642186, -0.08263041, -0.33069563,
  0.67456641, -1.19146982,  1.46848136,  1.99883652, -0.12777168,  0.36130379,
  0.04328583]
b = 4.11
```

## 3Ô∏è‚É£ Logistic Regression Model

- **Sigmoid function:** 

```python
def sigmoid(X, w, b):
    z = np.dot(X,w) + b
    f_wb = 1/(1+np.exp(-z))
    return f_wb
    
```


## 5Ô∏è‚É£ Predictions & Decision Threshold

- Predict probability of heart disease for new patients.  
- Use threshold `0.5` for binary decision:  
- `>= 0.5` ‚Üí Heart Disease  
- `< 0.5` ‚Üí No Heart Disease

```python
import numpy as np

x = np.array([
    80,   
    1,    
    3,  
    200,  
    300,  
    1,    
    2,    
    150,  
    1,    
    4.0, 
    2,   
    3,   
    3   
])
x_norm = (x - X_mean) / X_std

pred = sigmoid(x_norm, w, b)

# decision
decision = "Heart Disease" if pred >= 0.5 else "No Heart Disease"

print(f"üíì Heart Disease Probability: {pred*100:.2f}% ‚Üí Prediction: {decision}")
```

## 4Ô∏è‚É£ Model Training & Cost History

- Track cost at each iteration.  
- Plot cost reduction to ensure proper training.

**Visualization Idea:**  
- Cost vs iterations line chart to spot convergence issues or exploding gradients.

---


```python
def compute_cost(X,y,w,b):
    m = X.shape[0]
    epsilon = 1e-8
    cost = 0.0
    f_wb = sigmoid(X,w,b)
    cost = -(1/m)*np.sum(y*np.log(f_wb + epsilon) + (1-y)*np.log(1-f_wb + epsilon))
    return cost
    
```

```python
def compute_gradient(X,y,w,b):
    m = X.shape[0]
    err = sigmoid(X,w,b) - y 
    dj_dw = (1/m)*np.dot(X.T, err)
    dj_db = (1/m)*np.sum(err)
    return dj_db, dj_dw
```

```python
X_tmp = np.array([[0.5, 1.5], [1,1], [1.5, 0.5], [3, 0.5], [2, 2], [1, 2.5]])
y_tmp = np.array([0, 0, 0, 1, 1, 1])
w_tmp = np.array([2.,3.])
b_tmp = 1.
dj_db_tmp, dj_dw_tmp = compute_gradient(X_tmp, y_tmp, w_tmp, b_tmp)
print(f"dj_db: {dj_db_tmp}" )
print(f"dj_dw: {dj_dw_tmp.tolist()}" )
```

```python
def gradient_descent(X, y, w_in, b_in, cost_function, gradient_function, alpha, num_iters): 
    
    J_history = []
    w = copy.deepcopy(w_in)  
    b = b_in
    
    for i in range(num_iters):

        # Calculate the gradient and update the parameters
        dj_db,dj_dw = gradient_function(X, y, w, b)   ##None

        # Update Parameters using w, b, alpha and gradient
        w = w - alpha * dj_dw               ##None
        b = b - alpha * dj_db               ##None
      
        # Save cost J at each iteration
        if i<100000:      # prevent resource exhaustion 
            J_history.append( cost_function(X, y, w, b))

        # Print cost every at intervals 10 times or as many iterations if < 10
        if i% math.ceil(num_iters / 10) == 0:
            print(f"Iteration {i:4d}: Cost {J_history[-1]:8.2f}   ")
        
    return w, b, J_history #return final w,b and J history for graphing
```

```python
# initialize parameters
initial_w = np.zeros(X_norm.shape[1])
initial_b = 0
# some gradient descent settings
alpha = 0.001  # Much smaller for unscaled features
iterations = 100000  # More iterations needed
# run gradient descent 
w_final, b_final, J_hist = gradient_descent(X_train, y_train, initial_w, initial_b,
                                                    compute_cost, compute_gradient, 
                                                    alpha, iterations)
print(f"b,w found by gradient descent: {b_final:0.2f},{w_final} ")
for i in range(X_test.shape[0]):
    pred = np.dot(X_test.iloc[i], w_final) + b_final
    print(f"Test prediction: {pred:0.2f}, target value: {y_test.iloc[i]}")


```

## 6Ô∏è‚É£ Visualizations

- Probabilities bar chart for multiple patients.  
- Color-coded for risk levels.

---

```python
sns.set(style="whitegrid")

# --- 1Ô∏è‚É£ Plot Cost History ---
plt.figure(figsize=(8,5))
plt.plot(J_hist, color='blue')
plt.title("Cost Function Convergence", fontsize=14)
plt.xlabel("Iterations")
plt.ylabel("Cost")
plt.show()

# --- 2Ô∏è‚É£ Predict on Test Set & visualize probability ---
y_pred_probs = sigmoid(X_test, w_final, b_final)  # probabilities
y_pred_labels = (y_pred_probs >= 0.5).astype(int)

# Combine actual & predicted probabilities
df_pred = pd.DataFrame({
    "Actual": y_test,
    "Pred_Prob": y_pred_probs,
    "Pred_Label": y_pred_labels
})

# --- 3Ô∏è‚É£ Probability Distribution Plot ---
plt.figure(figsize=(8,5))
sns.histplot(df_pred['Pred_Prob'], bins=20, kde=True, color='red')
plt.title("Predicted Heart Disease Probabilities", fontsize=14)
plt.xlabel("Probability")
plt.ylabel("Count")
plt.show()

# --- 4Ô∏è‚É£ Actual vs Predicted Scatter Plot ---
plt.figure(figsize=(8,5))
plt.scatter(range(len(y_test)), y_test, label="Actual", color='blue', alpha=0.6)
plt.scatter(range(len(y_test)), y_pred_probs, label="Predicted Prob.", color='red', alpha=0.6)
plt.title("Actual vs Predicted Probability", fontsize=14)
plt.xlabel("Test Sample Index")
plt.ylabel("Probability")
plt.legend()
plt.show()

# --- 5Ô∏è‚É£ Color-coded risk bar chart ---
risk_colors = ['green' if p<0.5 else 'orange' if p<0.8 else 'red' for p in y_pred_probs]

plt.figure(figsize=(10,5))
plt.bar(range(len(y_pred_probs)), y_pred_probs, color=risk_colors)
plt.title("Heart Disease Risk Levels", fontsize=14)
plt.xlabel("Patient Index")
plt.ylabel("Predicted Probability")
plt.show()
```
